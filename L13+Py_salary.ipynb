```python
from sklearn.neighbors import KNeighborsClassifier
import networkx as nx
import pandas as pd
from collections import defaultdict
import random
import numpy as np
# 数据加载
file = 'jobs_4k.xls'
data = pd.read_excel(file)
# data.head(5)
position_name = data['positionName'].tolist()
skill_lables = data['skillLables'].tolist()
position_skill_graph = defaultdict(list)
for p,s in zip(position_name,skill_lables):
    position_skill_graph[p] += eval(s)
    
# print(position_skill_graph)
import matplotlib.pyplot as plt
G = nx.Graph(position_skill_graph)
# 设置字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 抽取
sample_nodes = random.sample(position_name,k =3)
sample_nodes_connection = sample_nodes
for p,skill in position_skill_graph.items():
    if p in sample_nodes:
        sample_nodes_connection += skill
sample_graph = G.subgraph(sample_nodes_connection)
plt.figure(figsize = (50,30))
pos = nx.spring_layout(sample_graph,k = 1)
nx.draw(sample_graph,pos,with_lables = True,node_size = 30,font_size = 10)
# plt.show()

# PageRank排序
pr = nx.pagerank(G,alpha = 0.9)
ranked_position_and_skill = sorted( [(name,value) for name,value in pr.items()],key = lambda x: x[1],reverse = True)
print(ranked_position_and_skill)
```

```python
X_content = data.drop(['salary'],axis = 1)
target = data['salary']
X_content['merged'] = X_content.apply(lambda x: ''.join(str(x)),axis = 1)

# print(X_content['merged'][0])
```

```python
X_string = X_content['merged'].tolist()
# print(X_string)


def get_one_row_job_string(x_string_row):
    job_string = ''
    for i , element in enumerate(x_string_row.split('\n')):
        if len(element.split()) == 2:
            _,value = element.split()
            if i == 0:
                continue
            job_string += value
    return job_string
  
```

```python
import jieba
import re
def token(string):
    return re.findall('\w+',string)

cutted_X = []
for i,row in enumerate(X_string):
#     print(row)
    
    job_string = get_one_row_job_string(row)
#     print(job_string)
    
    cutted_X.append(' '.join(list(jieba.cut(''.join(token(job_string)) ) ) ) )  

# print(cutted_X)
   
```

```python
# 特征提取
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(cutted_X)
target_num = [np.mean( list( map(float,re.findall('\d+',s) ) ) ) for s in target]
# print(target_num)
y = target_num
```

```python
# KNN模型
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor( n_neighbors = 3 )
model.fit(X,y)
```

```python
# 测试模型

def predict_test(test_string,model):
    test_word = list(jieba.cut(test_string))
    test_vector = vectorizer.transform(test_word)
    y_pred = model.predict(test_vector)
    return y_pred[2]
persons = [ '北京本科四年算法工程师',
            '深圳硕士两年数据分析师',
            '青岛本科四年大数据工程师']
for p in persons:
    print( "{}薪资预测为{}".format(p,predict_test(p,model)))
```
