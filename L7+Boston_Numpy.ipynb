```python
import numpy as np
from sklearn.datasets import load_boston
```

```python
# 数据加载/特征提取和Label转换
data = load_boston()
x = data['data']
y = data['target']
y = y.reshape(-1,1)
```

```python
# 数据规范化-正态分布
x = (x - np.mean(x,axis = 0 ))/np.std(x,axis = 0 )
n_feature = x.shape[1]
print(n_feature)
```

```python
# 三层网络结构,一层隐藏层
n_hidden = 10
```

```python
# 参数随机初始化
w1 = np.random.randn(n_feature,n_hidden)
b1 = np.zeros(n_hidden)
w2 = np.random.randn(n_hidden,1)
b2 = np.zeros(1)
```

```python
# 定义relu,设置学习率
def Relu(x):
    return np.where(x<0,0,x)
learning_rate = 1e-6
```

```python
# 定义Loss
def MSE_Loss(y,y_hat):
    result = np.mean(np.square(y_hat - y))
    return result
# 定义函数
def Linear(x,w1,b1):
    result = x.dot(w1) + b1
    return result
```

```python
# 定义循环
for t in range(5000):
#     前向传播
    l1 = Linear(x,w1,b1)
    s1 = Relu(l1)
    y_pred = Linear(s1,w2,b2)
    loss = MSE_Loss(y,y_pred)
    print(t,loss)
#     反向传播
    grad_y_pred = 2.0*(y_pred - y)
    grad_w2 = s1.T.dot(grad_y_pred)
    grad_temp_relu = grad_y_pred.dot(w2.T)
    grad_temp = grad_temp_relu.copy()
    grad_temp_relu[l1< 0 ] = 0
    grad_w1 = x.T.dot(grad_temp_relu)
    # 权重更新
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2
print('w1={} \n w2={}'.format(w1,w2))
```

```python
学习率设置
learning_rate = 1e-4 不收敛
learning_rate = 1e-6 结果不如1e-5
```
